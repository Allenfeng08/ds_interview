```python
# longest substring without repeating characters
# time complexity:
# space complexity:

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        max_len = 0
        left = 0
        stack = []

        for i in range(0, len(s)):
            while s[i] in stack:
                stack.pop(0)
                left += 1
            stack.append(s[i])
            max_len = max(max_len, i-left+1)

        return max_len
```

```python
test = Solution()
test.lengthOfLongestSubstring("abcabcbb")
```

```python
# longest palindromic substring (leetcode 5)
# time complexity:
# space complexity:

class Solution:
    def longestPalindrom(self, s: str) -> str:
        if not s or len(s) == 1:
            return s
        
        start, end = 0, 0

        def expand(left, right):
            nonlocal start, end
            while left >= 0 and right <= len(s)-1 and s[left] == s[right]:
                left -= 1
                right += 1
            if (right-1) - (left+1) + 1 > end - start + 1:
                start = left + 1
                end = right - 1
        
        for i in range(len(s)):
            expand(i, i)
            expand(i, i+1)
        
        return s[start:end+1]
```

```python
test = Solution()
test.longestPalindrom("ababadefg")
```

```python
# plus one
# example:
# [1,2,3] -> [1,2,4]
# [9] -> [1,0]

class Solution:
    def plus_one(self, digits: list) -> list:
        if not digits:
            return [1]
        
        carry = 1
        result = []

        for i in range(-1, -len(digits)-1, -1):
            add = carry + digits[i]
            carry = add//10
            remain = add%10
            result.append(remain)
        if carry > 0:
            result.append(carry)
        return result[::-1]
```

```python
# add binary
# example:
# a = '11', b = '1', return '100'
# a = '1010', b = '1011', return '10101'

class Solution:
    def add_binary(self, a: str, b: str) -> str:
        carry = 0
        result = []
        i = len(a)-1
        j = len(b)-1
        while carry > 0 or i>=0 or j>=0:
            add = carry
            if i>=0:
                add += int(a[i])
                i-=1
            if j>=0:
                add += int(b[j])
                j-=1
            carry = add//2
            result.append(str(add%2))
        
        return ''.join(result[::-1])
```

```python
test = Solution()
test.add_binary('1010','1011')
```

```python
# sum of two
# example:
# nums = [2,1,4,3,5,6,7,7], target = 8, return [[1,7],[2,6],[3,5]]

class Solution:
    def sum_of_two(self, nums: list, target: int) -> list:
        if not nums:
            return []
        
        nums.sort()
        i = 0
        j = len(nums)-1
        result = []
        
        while i < j:
            add = nums[i] + nums[j]
            if add == target:
                result.append([nums[i], nums[j]])
                left = nums[i]
                right = nums[j]
                while i<j and nums[i] == left:
                    i+=1
                while i<j and nums[j] == right:
                    j-=1
            elif add < target:
                i+=1
            else:
                j-=1
        return result
        
```

```python
test = Solution()
test.sum_of_two([2,1,4,3,5,6,7,7], 8)
```

```python
# find minimal peak in an array
# example: 
# arr = [2,7,8,5,1,6,3,9,4], return 6

class Solution:
    def find_minimal_peak(self, arr: list) -> list:
        min_peak = float('inf')
        min_idx = None

        if not arr:
            return []
        if len(arr) == 1:
            return [arr[0], 0]
        for i in range(len(arr)):
            if i == 0:
                if arr[i] > arr[i+1] and arr[i] < min_peak:
                    min_peak, min_idx = arr[i], i
            elif i == len(arr)-1:
                if arr[i] > arr[i-1] and arr[i] < min_peak:
                    min_peak, min_idx = arr[i], i
            else:
                if arr[i] > arr[i-1] and arr[i] > arr[i+1] and arr[i] < min_peak:
                    min_peak, min_idx = arr[i], i
        return [min_peak, min_idx]
```

```python
test = Solution()
test.find_minimal_peak([2,7,8,5,1,6,3,9,4])
```

```python
# calculate the power
# example:
# a = 2, b = 0, return 1
# a = 2, b = -3, return 0.125
# a = 2, b = 10, return 1024

class Solution:
    def power(self, a: int, b: int) -> float:
        if b == 0:
            return 1
        elif b < 0:
            return 1/self.power(a, -b)
        else:
            if b%2 == 0:
                return self.power(a*a, b//2)
            else:
                return a*self.power(a, b-1)
```

```python
test = Solution()
test.power(2, 10)
```

```python
# calculate n factorial

class Solution:
    # top down
    def factorial_top_down(self, n: int) -> int:
        memo = {}
        def factorial(n, memo):
            if n == 0 or n == 1:
                result = 1
            elif n in memo:
                result = memo[n]
            else:
                result = n*factorial(n-1, memo)
            memo[n] = result
            return result
        return factorial(n, memo)
    
    # bottom up
    def factorial_bottom_up(self, n: int) -> int:
        if n == 0 or n == 1:
            return 1
        memo = []
        memo.append(1)
        memo.append(1)
        for i in range(2, n+1):
            next = i*memo[-1]
            memo.append(next)
        return memo[-1]
```

```python
test.factorial_bottom_up(5)
```

```python
class Solution:
    def fib_top_down(self, n: int) -> int:
        memo = {}
        def fib(n, memo):
            if n == 0:
                result = 0
            elif n == 1:
                result = 1
            elif n in memo:
                result = memo[n]
            else:
                result = fib(n-1, memo) + fib(n-2, memo)
            return result
        return fib(n, memo)
    
    def fib_bottom_up(self, n: int) -> int:
        if n == 0:
            return 0
        if n == 1:
            return 1
        memo = []
        memo.append(0)
        memo.append(1)

        for i in range(2, n+1):
            memo.append(memo[-1]+memo[-2])
        return memo[-1]
```

```python
test.fib_bottom_up(10)
```

```python
# sum of non-zero digits of an integer

class Solution:
    def sum_non_zero_digits_sol_one(self, n: int) -> int:
        if not n:
            return 0
        result = 0
        n = abs(n)
        while n > 0:
            digit = n%10
            if digit != 0:
                result += digit
            n //= 10
        return result
    
    def sum_non_zero_digits_sol_two(self, n: int) -> int:
        return sum(map(int, list(str(n))))
```

```python
# find max and min in a list (length n) with 1.5*n comparisons maximum

class Solution:
    def find_min_max(self, arr: list):
        greater = []
        less = []
        n_pair = len(arr)//2

        for i in range(0, 2*n_pair, 2):
            if arr[i+1] > arr[i]:
                greater.append(arr[i+1])
                less.append(arr[i])
            else:
                greater.append(arr[i])
                less.append(arr[i])
        if len(arr)%2 == 1:
            greater.append(arr[-1])
            less.append(arr[-1])
        
        min_val = float('inf')
        max_val = -float('inf')

        for val in greater:
            if val > max_val:
                max_val = val

        for val in less:
            if val < min_val:
                min_val = val

        return [min_val, max_val]
```

```python
test = Solution()
test.find_min_max([1,5,7,3,2,4,5,6,7,12,3,4,5,4,3])
```

```python
# # Find the word with the minimum distance using insertions, deletions, or substitutions (dynamic programming)

class Solution:
    def edit_distance(self, s1, s2):
        if len(s1) < len(s2):
            return self.edit_distance(s2, s1)
        if not s2:
            return len(s1)

        previous_row = range(len(s2) + 1)
        for i, c1 in enumerate(s1):
            current_row = [i + 1]
            for j, c2 in enumerate(s2):
                insertion = current_row[j] + 1
                deletion = previous_row[j + 1] + 1
                substitution = previous_row[j] + (c1 != c2)
                current_row.append(min(insertion, deletion, substitution))
            previous_row = current_row
        return previous_row[-1]

    def min_edit_distance_word(self, target, wordlist):
        return min(wordlist, key=lambda w: self.edit_distance(target, w))
```

```python
test = Solution()
test.min_edit_distance_word("hold", ["happy", "help", "hash", "coding", "watn", "offer"])
```

```python
# sorting

class sorting:
    # selection sort
    # time complexity: O(n^2)
    def find_min(self, arr: list) -> list:
        curr_min = arr[0]
        curr_min_idx = 0

        for i in range(1, len(arr)):
            if arr[i] < curr_min:
                curr_min = arr[i]
                curr_min_idx = i
        return [curr_min, curr_min_idx]
    
    def selection_sort(self, arr: list) -> list:
        result = []
        while arr:
            curr_min, curr_min_idx = self.find_min(arr)
            result.append(curr_min)
            arr.pop(curr_min_idx)
        return result
    
    # quick sort:
    # divide and conquer
    # choose pivot and divide the array into two parts: greater than pivot value and smaller than pivot value
    # time complexity: O(nlog(n)) on average, worst case is O(n^2)
    # in worst case senario, the pivot is always the largest or smallest number
    def quick_sort(self, arr: list) -> list:
        if not arr or len(arr) == 1:
            return arr
        pivot = arr[0]
        less = [num for num in arr[1:] if num <= pivot]
        greater = [num for num in arr[1:] if num > pivot]

        return self.quick_sort(less) + [pivot] + self.quick_sort(greater)
    
    # merge sort
    # divide and conquer
    # time complexity O(nlog(n))
    # recursively divide the array into two halves, sort the two halves, merge operation
    def merge_sort(self, arr: list) -> list:
        if len(arr) > 1:
            mid_idx = len(arr)//2
            sorted_left = self.merge_sort(arr[:mid_idx])
            sorted_right = self.merge_sort(arr[mid_idx:])

            i = 0
            j = 0
            k = 0

            while i < len(sorted_left) and j < len(sorted_right):
                if sorted_left[i] <= sorted_right[j]:
                    arr[k] = sorted_left[i]
                    i += 1
                    k += 1
                else:
                    arr[k] = sorted_right[j]
                    j += 1
                    k += 1
            if i < len(sorted_left):
                arr[k:] = sorted_left[i:]
            if j < len(sorted_right):
                arr[k:] = sorted_right[j:]
        return arr
```

```python
# path to climb stairs with minimum cost
# n: total number of stairs
# k: max steps allowed for each climb, k<=n
# p: price of stepping on each step, len(p)=n+1, starting from the ground (0)
# ex: n = 8, k = 2, p = [0, 3, 2, 4, 6, 1, 1, 5, 3]

class Solution:
    def path_of_climbing_with_min_cost(self, n: int, k: int, p: int) -> list:
        total_price = []
        pre_stair = []

        for i in range(0, k+1, 1):
            total_price.append(p[i])
            pre_stair.append(0)

        if n > k:
            for i in range(k+1, n+1, 1):
                candidate_pre_stair_idx = list(range(i-1, i-k-1, -1))
                smaller_price = float('inf')
                target_pre_stair_idx = None

                for idx in candidate_pre_stair_idx:
                    if total_price[idx] < smaller_price:
                        smaller_price = total_price[idx]
                        target_pre_stair_idx = idx
                total_price.append(smaller_price + p[i])
                pre_stair.append(target_pre_stair_idx)

        m = n
        min_cost_path = []
        min_cost_path.append(m)
        while m > 0:
            min_cost_path.append(pre_stair[m])
            m = pre_stair[m]
        return min_cost_path[::-1]
```

```python
test = Solution()
test.path_of_climbing_with_min_cost(8, 2, [0,3,2,4,6,1,1,5,3])
```

```python
# given a number n, count the total occurance of 0, 2, and 4 in the list from [0, 1, 2, ..., n]
# for example, when n == 22, the results is 11
# this is because the count for 0, 2, and 4 is {0:3, 2:6, 4:2}

class Solution:
    def count_num(self, n):
        import numpy as np
        num_arr = np.arange(n+1)
        count = {}
        for num in num_arr:
            for digit in str(num):
                if digit in count:
                    count[digit] += 1
                elif digit=="0" or digit=="2" or digit=="4":
                    count[digit] = 1
                else:
                    continue

        return sum(count.values())
```

```python
test = Solution()
test.count_num(22)
```

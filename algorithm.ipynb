```python
# longest substring without repeating characters
# time complexity:
# space complexity:

class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        max_len = 0
        left = 0
        stack = []

        for i in range(0, len(s)):
            while s[i] in stack:
                stack.pop(0)
                left += 1
            stack.append(s[i])
            max_len = max(max_len, i-left+1)

        return max_len
```

```python
test = Solution()
test.lengthOfLongestSubstring("abcabcbb")
```

```python
# longest palindromic substring
# time complexity:
# space complexity:

class Solution:
    def longestPalindrom(self, s: str) -> str:
        if not s or len(s) == 1:
            return s
        
        start, end = 0, 0

        def expand(left, right):
            nonlocal start, end
            while left >= 0 and right <= len(s)-1 and s[left] == s[right]:
                left -= 1
                right += 1
            if (right-1) - (left+1) + 1 > end - start + 1:
                start = left + 1
                end = right - 1
        
        for i in range(len(s)):
            expand(i, i)
            expand(i, i+1)
        
        return s[start:end+1]
```

```python
test = Solution()
test.longestPalindrom("ababadefg")
```

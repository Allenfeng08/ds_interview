```python
# Given a string, find all bigram and their counts

def bigram_count(string):
    string_list = string.split()
    d = {}
    for i in range(len(string_list)-1):
        if tuple(string_list[i:i+2]) in d:
            d[tuple(string_list[i:i+2])] += 1
        else:
            d[tuple(string_list[i:i+2])] = 1
    return d

string = "the quick person did not realize his speed and the quick person bumped "
bigram_count(string)
```

```python
# Find the two largest numbers in a array of integers

def find_two_largest(arr):
    import numpy as np
    largest = -np.inf
    second_largest = -np.inf

    for item in arr:
        if item > largest:
            second_largest = largest
            largest = item
        elif item > second_largest:
            second_largest = item
    return [largest, second_largest]
```

```python
# Fibonacci with memo - bottom up

def fib(n):
    if n == 1 or n == 2:
        return 1
    memo = []
    memo.append(1)
    memo.append(1)
    for i in range(2,n):
        next = memo[i-2]+memo[i-1]
        memo.append(next)
    return memo[-1]

fib(10)
```

```python
# Fibonacci with memo - top down

def fib_with_memo(n):
    memo = {}
    def fib(n, memo):
        if n==1 or n==2:
            result = 1
        elif n in memo:
            result = memo[n]
        else:
            result = fib(n-1, memo) + fib(n-2, memo)
        memo[n] = result
        return result
    return fib(n, memo)

fib_with_memo(10)
```

```python
# 0-1 knapsack problem
# description: give a list of items with weights and corresponding values
# you have a container with a weight limit
# select the items such that the total weights within the limit of container and the total value is maximized
# both time and space complexity is O(n * w_max)
```

```python
# dynamic programming

def knapsack(w, v, w_max, n):
    dp = [[0 for x in range(w_max+1)] for x in range(n+1)]

    for i in range(n+1):
        for j in range(w_max+1):
            if i == 0 or j == 0:
                dp[i][j] = 0
            elif w[i-1] <= j:
                dp[i][j] = max(dp[i-1][j], v[i-1]+dp[i-1][j-w[i-1]])
            else:
                dp[i][j] = dp[i-1][j]
    return dp[n][w_max]


w = [10, 20, 30]
v = [60, 100, 120]
w_max = 50
n = len(v)
print(knapsack(w, v, w_max, n))
```

```python
# Minimize cost of painting N houses such that adjacent houses have different colors
# Given an integer N and a 2D array cost, where cost[i][0], cost[i][1], and cost[i][2] is the cost of painting ith house with colors red, blue, and green respectively
# Find the minimum cost to paint all the houses such that NO two adjacent houses have the same color

# time-complexity: O(n), space-complexity: O(1)

def minCostToPaintHouses(cost):
    if not cost:
        return 0

    # Copy the first house's cost
    dp = cost[0]

    # Iterate over each house starting from the second one
    for i in range(1, len(cost)):
        # Current house cost when painted with red, blue, or green
        curr_red = cost[i][0] + min(dp[1], dp[2])
        curr_blue = cost[i][1] + min(dp[0], dp[2])
        curr_green = cost[i][2] + min(dp[0], dp[1])

        # Update dp with the current house's costs
        dp = [curr_red, curr_blue, curr_green]

    # The minimum cost will be the smallest cost of painting the last house
    return min(dp)


# Example usage:
cost = [[17, 2, 17], [16, 16, 5], [14, 3, 19], [2, 3, 8]]
print(minCostToPaintHouses(cost))
```

```python
# TSP
# nearest insertion

import numpy as np

def find_nearest_unvisited(city, unvisited, distance_matrix):
    # Find the nearest unvisited city from the current city
    nearest_city = min(unvisited, key=lambda x: distance_matrix[city][x])
    return nearest_city, distance_matrix[city][nearest_city]

def find_insertion_position(city, tour, distance_matrix):
    # Find the position in the tour where inserting the city causes the least increase in length
    best_increase = float('inf')
    best_position = None
    for i in range(len(tour) - 1):
        current_increase = (distance_matrix[tour[i]][city] + 
                            distance_matrix[city][tour[i+1]] - 
                            distance_matrix[tour[i]][tour[i+1]])
        if current_increase < best_increase:
            best_increase = current_increase
            best_position = i + 1
    return best_position

def nearest_insertion(distance_matrix):
    # Start tour at the first city
    tour = [0, 0]  # start and end at the first city to make it a cycle
    unvisited = set(range(1, len(distance_matrix)))

    # Build the tour
    while unvisited:
        # Find the nearest unvisited city to the tour
        nearest_distance = float('inf')
        for city in tour:
            nearest_city, distance = find_nearest_unvisited(city, unvisited, distance_matrix)
            if distance < nearest_distance:
                nearest_distance = distance
                next_city = nearest_city

        # Find the best position to insert the next city
        insertion_position = find_insertion_position(next_city, tour, distance_matrix)
        tour.insert(insertion_position, next_city)
        unvisited.remove(next_city)

    # Calculate total cost of the tour
    total_cost = sum(distance_matrix[tour[i]][tour[i+1]] for i in range(len(tour)-1))

    return total_cost, tour
```

```python
# Example distance matrix
distance_matrix = np.array([
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
])

cost, tour = nearest_insertion(distance_matrix)
print(f"Minimum cost: {cost}")
print(f"Tour: {tour}")
```

```python
# generate random number by weights
# similar to LC528

def weighted_choice(numbers, weights):
    total_sum = 0
    cum_sum = []
    for weight in weights:
        total_sum += weight
        cum_sum.append(total_sum)
    
    import random
    target = total_sum*random.random()
    
    left = 0
    right = len(numbers)

    while left <= right:
        mid = (left+right)//2
        if cum_sum[mid] == target:
            return numbers[mid]
        elif cum_sum[mid] > target:
            right = mid-1
        else:
            left = mid+1
    return numbers[left]


# Example usage:
numbers = [10, 20, 30, 40, 50]
weights = [1, 2, 3, 4, 5]

weighted_choice(numbers, weights)
```

```python
# use dice to generate uniform distribution between 1 to 10
# rejection sampling

import random

def roll_die():
    return random.randint(1, 6)

def generate_uniform_distribution(lower, upper):
    roll_again = True
    while roll_again:
        first_roll = roll_die()-1
        second_roll = roll_die()-1    
        number = first_roll*6 + second_roll # 0 - 35
        if number >= lower and number <= upper:
            return number + 1
```

```python
# selection sort
# find the smallest of an array
# pop the element and find the smallest of the rest
# time-complexity O(n^2)
# space-complexity O(1) for in-place
```

```python
def find_smallest(arr):
    import numpy as np
    smallest_num = arr[0]
    smallest_idx = 0
    for idx in np.arange(1,len(arr)):
        if arr[idx] < smallest_num:
            smallest_num = arr[idx]
            smallest_idx = idx
    return [smallest_num, smallest_idx]

def selection_sort(arr):
    if len(arr) <= 1:
        return arr
    sorted_arr = []
    while arr:
        sorted_arr.append(find_smallest(arr)[0])
        arr.pop(find_smallest(arr)[1])
    return sorted_arr
```

```python
# quick sort
# divide and conquer
# choose pivot and divide the array into two parts: greater than pivot value and smaller than pivot value
# time complexity: O(nlog(n)) on average, worst case is O(n^2)
# in worst case senario, the pivot is always the largest or smallest number
# space complexity: O(log(n)) on average for in-place, O(n) worst for in-place
```

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    else:
        pivot = arr[0]
        less = [i for i in arr[1:] if i <= pivot]
        greater = [i for i in arr[1:] if i > pivot]
        return quick_sort(less) + [pivot] + quick_sort(greater)
```

```python
quick_sort([6,7,1,2,3,4,5,8,9,4,5])
```

```python
# merge sort
# divide and conquer
# time complexity O(nlog(n))
# space complexity O(n)
# recursively divide the array into two halves
# sort the two halves
# merge operation
# //: floor division
# %: modulus
```

```python
def merge_sort(arr):
    if len(arr)<=1:
        return arr
    else:
        mid = len(arr)//2
        sorted_left = merge_sort(arr[:mid])
        sorted_right = merge_sort(arr[mid:])

        i, j, k = 0, 0, 0
        while i < len(sorted_left) and j < len(sorted_right):
            if sorted_left[i] <= sorted_right[j]:
                arr[k] = sorted_left[i]
                i+=1
                k+=1
            else:
                arr[k] = sorted_right[j]
                j+=1
                k+=1
        if i < len(sorted_left):
            arr[k:] = sorted_left[i:]
        else:
            arr[k:] = sorted_right[j:]
    return arr
            
```

```python
# max running sum
def max_running_sum(arr):
    import numpy as np
    max_sum = -np.inf
    curr_sum = 0

    for i in range(len(arr)):
        curr_sum += arr[i]
        if curr_sum > max_sum:
            max_sum = curr_sum
    return max_sum


arr = [1, 2, 3, -4, -1, 3, 4, -1, -2]
max_running_sum(arr)
```

```python
# Find the all the factors of a integer and calculate the sum 
def find_factors(num):
    left = 1
    right = num
    result = []

    while left < right:
        if num%left == 0:
            result.append(left)
            if num//left != left:
                result.append(num//left)
            right = num//left
        left = left+1
    return sorted(result)

find_factors(88)
```

```python
def find_factors(num):
    import numpy as np
    factors = []
    sqrt_num = int(np.sqrt(num))
    
    for i in range(1, sqrt_num + 1):
        if num % i == 0:
            factors.append(i)
            if i != num // i:
                factors.append(num // i)
    
    return sorted(factors)

find_factors(88)
```

```python
# extract IP address
import pandas as pd
import re
from collections import Counter

df = pd.DataFrame({
    'IP': ['fskjfklsdjflks 127.0.0.1 fsjkfjl', 'fsjlkfjls. fksj 256.111.111.1', '123.123.123.123 fsjdlfk.sfsdf.fs..sfdfd..sf.fkjskf']
})

def is_valid_ip(word):
    parts = word.split('.')
    if len(parts) != 4:
        return False
    for part in parts:
        if not part.isdigit() or not 0 <= int(part) <= 255:
            return False
    return True

all_ips = []
for row in df['IP']:
    words = row.split()
    for word in words:
        if is_valid_ip(word):
            all_ips.append(word)

ip_count = Counter(all_ips)
sorted_ip_count = sorted(ip_count.items(), key=lambda x: x[1], reverse=True)
for ip, count in sorted_ip_count:
    print(f'IP Address: {ip}, Frequency: {count}')
```

```python
# 在所有整数中找出满足一个数等于这个数所有的digit的五次方相加，比如abcd=a^5+b^5+c^5+d^5，然后把所有满足条件的数加起来求和

def sum_of_fifth_powers(n):
    return sum(int(digit)**5 for digit in str(n))

def find_special_numbers():
    limit = 9**5 * 6
    special_numbers = []
    for i in range(2, limit):
        if i == sum_of_fifth_powers(i):
            special_numbers.append(i)
    return special_numbers, sum(special_numbers)

special_numbers, total_sum = find_special_numbers()
```

```python
# sqrt(x)
# binary search

def mySqrt(x):
    if x<0:
        return 'please input another integer that is greater than 0'
    if x==0:
        return 0
    if x==1:
        return 1
    
    left = 0
    right = x

    while left <= right:
        mid = (left + right)//2
        if mid*mid == x:
            return mid
        elif mid*mid > x:
            right = mid - 1
        else:
            left = mid + 1
    return right
```

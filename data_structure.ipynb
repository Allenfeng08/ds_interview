```python
# deque

from collections import deque
my_deque = deque([1,2,3])
```

```python
my_deque.append(4)
```

```python
my_deque
```

```python
my_deque.appendleft(0)
```

```python
my_deque
```

```python
my_deque.pop()
```

```python
my_deque
```

```python
my_deque.popleft()
```

```python
my_deque
```

```python
# rotate elements
my_deque.rotate(1) # shifts elements to the right by 1
```

```python
my_deque
```

```python
# zip

names = ["Alice", "Bob"]
scores = [85, 90]

for name, score in zip(names, scores):
    print(f"the score for {name}: {score}")
```

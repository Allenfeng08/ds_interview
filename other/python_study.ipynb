```python
class Person:
    def __init__(self, name, personality, isSitting, robot_owned):
        self.name = name
        self.personality = personality
        self.isSitting = isSitting
        self.robot_owned = robot_owned

    def sit_down(self):
        self.isSitting = True

    def stand_up(self):
        self.isSitting = False
```

```python
class tic_tac_toe:
    def __init__(self):
        self.board = [['-' for _ in range(3)] for _ in range(3)]
        self.taken = set()
        self.current_player = 'X'

    def switch_player(self):
        if self.current_player == 'X':
            self.current_player = 'O'
        else:
            self.current_player = 'X'

    def valid_selection(self, i, j):
        if (i,j) not in self.taken:
            if 0<=i<=2 and 0<=j<=2:
                self.taken.add((i,j))
                return True
        return False

    def selection(self, i, j):
        if self.valid_selection(i, j):
            self.board[i][j] = self.current_player
            self.switch_player()
        else:
            print('the selection is not valid, please make another selection')
    
    def check_winner(self):
        for i in range(3):
            if self.board[i][0] == self.board[i][1] == self.board[i][2] != '-':
                return print('the winner is {}'.format(self.board[i][0]))
            if self.board[0][i] == self.board[1][i] == self.board[2][i] != '-':
                return print('the winner is {}'.format(self.board[0][i]))
        if self.board[0][0] == self.board[1][1] == self.board[2][2] != '-':
            return print('the winner is {}'.format(self.board[0][0]))
        if self.board[0][2] == self.board[1][1] == self.board[2][0] != '-':
                return print('the winner is {}'.format(self.board[0][2]))
        if len(self.taken) == 9:
            return 'the game is draw'
        else:
            return 'no winner yet, game continue'
            
    def print_board(self):
        for row in self.board:
            print(' '.join(row))
    
    def start_game(self):
        while True:
            print('Player {} please make selection, the Current board: '.format(self.current_player))
            self.print_board()
            try:
                i = int(input('Please enter row index: '))
                j = int(input('Please enter column index: '))
                self.selection(i, j)
            except ValueError:
                print("Please enter valid integers for row and column indexes.")
                continue  # If input is invalid, continue to next iteration of loop

            result = self.check_winner()
            if result != 'no winner yet, game continue':
                break  # Exit the loop if there's a winner or a draw
```

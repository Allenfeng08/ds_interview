```python
# aggregation function

df_1 = df.groupby("planet_type").agg(
    unique_planet_volume = ("planet_name", "nunique"),
    total_planet_volume = ("planet_name", "size"),
    not_null_volume = ("planet_name", "count"),
    planet_radius_earth_avg = ("orbital_period_days", "mean"),
    planet_radius_earth_total = ("planet_radius_earth", "sum"),
    planet_radius_earth_min = ("planet_radius_earth", "min"),
    planet_radius_earth_max = ("planet_radius_earth", "max")
).reset_index().sort_values(by="planet_type", ascending=True)
```

```python
# window function: row number, rank, dense rank

df_3 = df.copy()
df_3["row_number"] = df_3["planet_radius_earth"].rank(method="first", ascending=True)
df_3["rank"] = df_3["planet_radius_earth"].rank(method="min", ascending=True)
df_3["dense_rank"] = df_3["planet_radius_earth"].rank(method="dense", ascending=True)
```

```python
# window function: row number, rank, dense rank - by group

df_4 = df.copy()
df_4["row_number"] = df_4.groupby("planet_type")["planet_radius_earth"].rank(method="first", ascending=True)
df_4["rank"] = df_4.groupby("planet_type")["planet_radius_earth"].rank(method="min", ascending=True)
df_4["dense_rank"] = df_4.groupby("planet_type")["planet_radius_earth"].rank(method="dense", ascending=True)
```

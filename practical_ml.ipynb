```python
# KNN
# given a list of lists called arr: [x, y, value], a target [x_0, y_0], and integer number k
# x: x coordinate
# y: y coordinate
# value: numerical value
# find the k nearest neighbours of target and sum up their values
```

```python
import numpy as np

class Solution:
    def knn(self, arr: list, target: list, k: int) -> float:
        distance = []
        for i in range(len(arr)):
            distance.append([np.sqrt((arr[i][0]-target[0])**2+(arr[i][1]-target[1])**2), arr[i][2]])
        
        distance_sorted = sorted(distance, key=lambda x: x[0])
        total = 0
        k = min(k, len(distance_sorted))
        for i in range(k):
            total += distance_sorted[i][1]
        return total
```

```python
test = Solution()
test.knn([[1,2,1],[2,1,2],[3,3,1]], [1,2], 2)
```

```python
# recommender system
# use pairwise deletion to handle None values
```

```python
import pandas as pd
import numpy as np
import random
import math
```

```python
class Solution:
    def cosine_similarity(self, arr_1, arr_2):
        np_arr_1 = np.array(arr_1, dtype=float)
        np_arr_2 = np.array(arr_2, dtype=float)
        mask = ~np.isnan(np_arr_1) & ~np.isnan(np_arr_2)
        if not np.any(mask):
            return np.nan
        a = np_arr_1[mask]
        b = np_arr_2[mask]
        norm_a = np.linalg.norm(a)
        norm_b = np.linalg.norm(b)
        if norm_a == 0 or norm_b == 0:
            return np.nan
        return np.dot(a,b)/(norm_a*norm_b)

    def customer_database_generator(self, customer_vol, dim):
        customer_database = []
        for _ in range(customer_vol):
            customer_database.append(random.choices([0,1,np.nan], weights=[0.4,0.4,0.2], k=dim))
        return customer_database

    def recommender_system(self, target_customer, customer_database, k):
        similarities = []
        
        for i in range(len(customer_database)):
            similarity = self.cosine_similarity(target_customer, customer_database[i])
            if not math.isnan(similarity):
                similarities.append([i, similarity])
        sorted_similarities = sorted(similarities, key=lambda x:x[1], reverse=True)
        
        idx = []
        k = min(k, len(similarities))
        for i in range(k):
            idx.append(sorted_similarities[i][0])
        
        top_customer = [customer_database[i] for i in idx]
        
        recommendations = []
        dim = len(target_customer)
        for i in range(dim):
            if math.isnan(target_customer[i]):
                if any(not math.isnan(top_customer[j][i]) and top_customer[j][i] == 1 for j in range(len(top_customer))):
                    recommendations.append(i)
        return recommendations
```

```python
test = Solution()
customer_database = test.customer_database_generator(200, 10)
test.recommender_system([1,0,np.nan,1,0,np.nan,1,0,np.nan,1], customer_database, 5)
```

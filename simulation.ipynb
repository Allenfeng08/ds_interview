```python
# use randint(1,5) to generate uniform distribution between 1 and 7 (randint(1,7))
# rejection sampling

import random

class Solution:
    def roll_die(self):
        return random.randint(1,5)
    
    def uniform_distribution_generator(self, lower, upper):
        roll_again = True
        while roll_again:
            first_roll = self.roll_die()-1
            second_roll = self.roll_die()-1
            gen_num = first_roll + second_roll*5 # 0 - 24
            if gen_num >= lower and gen_num <= upper:
                return gen_num
```

```python
test = Solution()

result = []
for _ in range(10000):
    result.append(test.uniform_distribution_generator(1,7))

from collections import Counter
Counter(result)
```

```python
# given a n-sided die, simulate how many times need to roll the die such that every side is shown
# coupon collector's problem

import random
import numpy as np

class Solution:
    def roll_die(self, n):
        return random.randint(1, n)
    def getCount(self, n):
        seen = set()
        cnt = 0
        while len(seen) < n:
            seen.add(self.roll_die(n))
            cnt += 1
        return cnt
    def getStats(self, m, n):
        result = []
        for _ in range(m):
            result.append(self.getCount(n))
        return [np.mean(result), np.median(result), np.std(result)]
    def draw_num_expectation(self, n):
        estimate = 0
        for i in range(1, n+1):
            estimate += n*(1/i)
        return estimate
```

```python
test = Solution()
test.getStats(10000, 10)
```

```python
test.draw_num_expectation(10)
```

```python
# calculate natural constant e without using built-in function
# e = 1/0! + 1/1! + 1/2! + ...

class Solution:
    def calculate_e(self, num_iter):
        estimate = 1
        factorial = 1

        for i in range(1, num_iter+1):
            factorial *= i
            estimate += 1/factorial
        return estimate
```

```python
test = Solution()
test.calculate_e(100)
```

```python
# given random.randint(0,1), create a random number between a and b

import random

class Solution:
    def roll_die(self):
        return random.randint(0, 1)

    def random_number_generator(self, a, b):
        if a > b:
            raise ValueError("a must be <= b")

        diff = b - a
        if diff == 0:
            return a

        power = 0
        temp = diff
        while temp > 0:
            temp //= 2
            power += 1

        while True:
            gen_num = 0
            for _ in range(power):
                gen_num = gen_num * 2 + self.roll_die()
            if gen_num <= diff:
                return gen_num + a
```

```python
test = Solution()
test.random_number_generator(2, 14)
```

```python
# one drunk passenger among 100 passengers
# what is the probability that the last passenger sit on the desigated sit
# the solution is 0.5

import random
import numpy as np
from collections import Counter

class Solution:
    def drunk_passenger(self, n):
        all_seat = set(np.arange(1,n+1))
        taken_seat = set()
        drunk_passenger_idx = random.randint(1,n)
        selection_seq = []
        for i in range(1,n+1):
            available_seat = all_seat - taken_seat
            if i == drunk_passenger_idx:
                selection = random.choice(list(available_seat))
            else:
                if i in available_seat:
                    selection = i
                else:
                    selection = random.choice(list(available_seat))
            selection_seq.append(selection)
            taken_seat.add(selection)
        return selection_seq[-1]
```

```python
test = Solution()
result = []
passenger_num = 100
sim_num = 100000

for _ in range(sim_num):
    result.append(test.drunk_passenger(passenger_num))

Counter(result)[passenger_num]/sim_num
```

```python
# two players toss a coin, whenever there is a "HT" sequence, people toss "T" wins
# A goes first, B goes second

import random

class Solution:
    def coin_toss(self):
        sequence = random.choices(["H", "T"], k=2)
        while sequence[-2:] != ["H", "T"]:
            sequence.append(random.choice(["H", "T"]))
        return len(sequence)%2 == 1
```

```python
test = Solution()
sim_num = 100000
result = []

for _ in range(sim_num):
    result.append(test.coin_toss())

result.count(True)/sim_num
```

```python
# A and B toss coins, if A gets HH, A wins; if B gets HT, B wins; together, tie
# the game continues until somebody wins
# calculate the probablity of winning for A and B

import random
from collections import Counter

class Solution:
    def coin_toss_two_seq(self):
        seq_a = []
        seq_b = []
        seq_a.append(random.choice(["H", "T"]))
        seq_b.append(random.choice(["H", "T"]))
        while True:
            seq_a.append(random.choice(["H", "T"]))
            seq_b.append(random.choice(["H", "T"]))
            if seq_a[-2:] == ["H", "H"]:
                if seq_b[-2:] == ["H", "T"]:
                    return "tie"
                return "A"
            if seq_b[-2:] == ["H", "T"]:
                return "B"
```

```python
test = Solution()
result = []
sim_num = 100000
for _ in range(sim_num):
    result.append(test.coin_toss_two_seq())

Counter(result)
```

```python
# each pack has n cards, m unique cards in total
# calculate the expected number of packs need to be purchase to collect all the cards

import random
import numpy as np

class Solution:
    def collect_card(self, n, m):
        i = 0
        collected = set()
        while len(collected) < m:
            i += 1
            drawn = set(random.sample(range(1, m+1), n))
            collected = collected.union(drawn)
        return i
```

```python
test = Solution()
n = 8
m = 151
sim_num = 10000
result = []

for _ in range(sim_num):
    result.append(test.collect_card(n, m))

np.mean(result)
```

```python
# monty hall
import random

class Solution:
    def monty_hall(self):
        items = ["goat", "goat", "car"]
        random.shuffle(items)

        player_pick_switch_false_idx = random.randint(0,2)
        player_pick_switch_false = items[player_pick_switch_false_idx]

        items.pop(player_pick_switch_false_idx)

        if items[0] == "goat" and items[1] == "goat":
            host_open_idx = random.randint(0,1)
        elif items[0] == "car" and items[1] == "goat":
            host_open_idx = 1
        else:
            host_open_idx = 0
        
        items.pop(host_open_idx)
        player_pick_switch_true = items[0]

        return [player_pick_switch_true, player_pick_switch_false]
```

```python
test = Solution()
sim_num = 10000
switch_true = []
switch_false = []

for _ in range(sim_num):
    result = test.monty_hall()
    switch_true.append(result[0])
    switch_false.append(result[1])
```
